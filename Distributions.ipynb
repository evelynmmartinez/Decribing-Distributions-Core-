Describing Distributions

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set_context('talk')
mpl.rcParams['figure.figsize'] = [12,6]
```

```python
pd.set_option('display.max_columns',100)
```

```python
df = pd.read_csv('Data/MedicalDataset.csv')
df.head()
```

## VitD_Levels

```python
sns.histplot(data=df, x='VitD_levels', kde=True)

mean_value = df['VitD_levels'].mean()
median_value = df['VitD_levels'].median()
std_value = df['VitD_levels'].std()

plt.axvline(mean_value, color='red', linestyle='--', label='Mean')
plt.axvline(median_value, color='green', linestyle='--', label='Median')

plt.axvline(mean_value + std_value, color='black', linestyle='--', label='+1 Std')
plt.axvline(mean_value - std_value, color='black', linestyle='--', label='-1 Std')

plt.axvspan(mean_value - std_value, mean_value + std_value, color='yellow', alpha=0.3, label='+/- 1 Std Range')

plt.legend()

plt.show()
```

- Is it Discrete or Continuous?

    Continuous.
    

- Does it have a skew? If so, which direction (+/-)

    No, it's very symmetric. 
    

- What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)

    Mesokurtic.
